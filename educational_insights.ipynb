```python

# This code is written in python
# Importing required modules
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
# The GWCutilities has functions to help format data printed to the console
import GWCutilities as util

# Read a comma separated values (CSV) files into a variable
# as a pandas DataFrame
lwd = pd.read_csv("livwell135.csv")

print("\n\n\nColumbia is a country in South America with a population of almost 50 million people, a quarter of whom are between the ages of 15 and 24.\n")
print("Since 1870, it has been the law there that everyone should have schooling up through 9th grade.\n")
print("So you might be surprised to find out that not all students have 9 years of schooling.\n")

input("Press return to continue.\n")

print("While exploring the Women's Well-being dataset, I found that in Colombia the number of years of schooling women had decreased as the number of people in the household increased. In fact, the average was less than 7 years for women from the largest households. ")

print("\n\nBased on these findings, my proposed research question is:\n")
print("What are the most important reasons women in larger households have fewer years of schooling, and what can be done to reverse this trend?")
countryMatchBooleanList = (lwd["country_name"] == "Colombia")
cd = lwd.loc[countryMatchBooleanList]

plt.scatter(cd["HD_women_size_mean"], cd["ED_educ_years_mean"], label="Colombia")

plt.legend(loc='upper right')
plt.title("Number of household members versus years of schooling")
plt.xlabel("Number of household members")
plt.ylabel("Years of schooling")

plt.show()
```
